# Distance entre la France et les autres pays

Source : Mayer, T. & Zignago, S. (2011)
Notes on CEPII’s distances measures : the GeoDist Database
CEPII Working Paper 2011-25

## Importation du fichier 'dist'

```python
# Importation
distance = pd.read_excel('dist_cepii.xls')
# Résultat
distance.head()
```

```python
# Dimension
distance.shape
```

```python
# Information
distance.info()
```

## Réorganisation du df

```python
# Suppression des colonnes inutiles
distance.drop(columns={'contig','comlang_off','comlang_ethno','colony','comcol','curcol','col45','smctry','distw','distwces'}, inplace=True)
# Résultat
distance.head()
```

```python
# Sélection des données qui concernent la France
distance = distance.loc[distance['iso_o']=='FRA']
# Résultat
distance.head()
```

## Importation du fichier 'geo'

```python
# Importation
geo = pd.read_excel('geo_cepii.xls')
# Résultat
geo.head()
```

```python
# Dimension
geo.shape
```

```python
# Information
geo.info()
```

## Réorganisation du df

```python
# pivot table
geo = geo[['iso3','area','continent','pays']]
# Résultat
geo.head()
```

## Jointure des deux df

```python
# jointure via la colonne 'iso3' et 'iso_d'
distanceFrance_monde = pd.merge(distance, geo, left_on='iso_d', right_on='iso3')
# Résultat
distanceFrance_monde.head()
```

# Coût procédure exportation

Source : https://donnees.banquemondiale.org/indicator/IC.EXP.CSDC.CD?end=2019&start=2014&view=chart

## Importation du fichier

```python
# Importation du fichier
coutExportation = pd.read_csv('cout_procedure_exportation.csv', sep=';')
# Résultat
coutExportation.head()
```

```python
# Dimension
coutExportation.shape
```

```python
# Information
coutExportation.info()
```

## Réorganisation du df

```python
# Pivot table 
coutExportation_2018 = coutExportation.pivot_table(index='Country Name', columns='Indicator Name', values='2018').reset_index()
# Résultat
coutExportation_2018.head()
```

# Création du df final

## Jointure de disponibiliteVolaille_2018 et population_2018

```python
# jointure via 'Zone'
df_export_2018 = pd.merge(disponibiliteVolaille_2018, population_2018, on='Zone')
# Résultat
df_export_2018.head()
```

```python
# Dimension
df_export_2018.shape
```

## Jointure de df_export_2018 avec securiteAlimentaire_2018

```python
# Jointure via 'Zone'
df_export_2018 = pd.merge(df_export_2018, securiteAlimentaire_2018, on='Zone')
# Résultat
df_export_2018.head()
```

```python
# Dimension
df_export_2018.shape
```

## Jointure de df_export_2018 et indicateurEconomique_2018

```python
# Jointure via 'Zone' et 'pays Name'
df_export_2018 = pd.merge(df_export_2018, indicateurEconomique_2018, left_on='Zone', right_on='pays Name', how='left')
# Résultat
df_export_2018.head()
```

```python
# Suppression de la colonne 'pays Name'
df_export_2018.drop(columns={'pays Name'}, inplace=True)
# Dimension
df_export_2018.shape
```

## Jointure de df_export_2018 et distanceFrance_monde

```python
# jointure via 'Zone' et 'pays'
df_export_2018 = pd.merge(df_export_2018, distanceFrance_monde, left_on='Zone', right_on='pays', how='left')
# Résultat
df_export_2018.head()
```

```python
# Suppression de la colonne 'pays'
df_export_2018.drop(columns={'pays'}, inplace=True)
# Dimension
df_export_2018.shape
```

## Jointure de df_export_2018 et coutExportation_2018

```python
# Jointure via 'Zone' et 'Country Name'
df_export_2018 = pd.merge(df_export_2018, coutExportation_2018, left_on='Zone', right_on='Country Name', how='left')
# Résultat
df_export_2018.head()
```

```python
# Suppression de la colonne 'Country Name'
df_export_2018.drop(columns={'Country Name'}, inplace=True)
# Dimension
df_export_2018.shape
```

# Nettoyage de df_export_2018

## Index

```python
# Mise en index de la colonne 'Zone'
df_export_2018 = df_export_2018.set_index('Zone')
```
